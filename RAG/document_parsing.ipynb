```python
from azure.core.credentials import AzureKeyCredential
from azure.ai.formrecognizer import DocumentAnalysisClient
import os
from dotenv import load_dotenv

load_dotenv()

endpoint = os.getenv("AZURE_D_I_ENDPOINT")
key = os.getenv("AZURE_D_I_KEY")

document_analysis_client = DocumentAnalysisClient(
    endpoint = endpoint,
    credential = AzureKeyCredential(key))

invoice_path = "sample-local-pdf.pdf"
with open(invoice_path, "rb") as invoice:
    poller = document_analysis_client.begin_analyze_document("prebuilt-read", document=invoice)
    result = poller.result()

```

```python

for page in result.pages:
    print(result)
```
